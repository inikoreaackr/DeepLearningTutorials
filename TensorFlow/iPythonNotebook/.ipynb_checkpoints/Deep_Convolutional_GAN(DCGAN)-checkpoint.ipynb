```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# MNIST   .
x_train = mnist.train.images
y_train = mnist.train.labels
x_test = mnist.test.images
y_test = mnist.test.labels

print("x_train Shape: ", x_train.shape)
print("y_train Shape: ", y_train.shape)
print("x_test Shape: ", x_test.shape)
print("y_test Shape: ", y_test.shape)

def plot_mnist(data, classes, row_len=5):
    
    for i in range(10):
        idxs = (classes == i) #  index True  False .
        #  i  10 
        images = data[idxs][0:10] #True  index data    10 .
            
        for j in range(row_len):   
            plt.subplot(row_len, 10, i + j*10 + 1) # , , plot
            plt.imshow(images[j].reshape(28, 28), cmap='gray')
            #   
            if j == 0:
                plt.title(i)
            plt.axis('off')
    plt.show()

classes = np.argmax(y_train, 1) # [7 3 4 ..., 5 6 8]  .
plot_mnist(x_train, classes)
```

```python
# Leaky ReLU 
def lrelu(x, leak=0.2, name="lrelu"):
    return tf.maximum(x, leak*x)

def conv2d(input_, output_dim, 
       k_h=5, k_w=5, d_h=2, d_w=2, stddev=0.02, name="conv2d"):
    with tf.variable_scope(name):
        w = tf.get_variable('w', [k_h, k_w, input_.get_shape()[-1], output_dim],
                  initializer=tf.truncated_normal_initializer(stddev=stddev))
        conv = tf.nn.conv2d(input_, w, strides=[1, d_h, d_w, 1], padding='SAME')

        biases = tf.get_variable('biases', [output_dim], initializer=tf.constant_initializer(0.0))
        conv = tf.reshape(tf.nn.bias_add(conv, biases), conv.get_shape())

    return conv
```

```python
num_classes = 2
input_size = 28*28
z_size = 7*7*64
num_classes = 2 #   
learning_rate = 1e-3
num_of_iter = 5000
batch_size = 100

weights_D = {
        'W_conv1': tf.Variable(tf.truncated_normal([5, 5, 1, 32], stddev=0.1), name='W_conv1_D'),
        'W_conv2': tf.Variable(tf.truncated_normal([5, 5, 32, 64], stddev=0.1), name='W_conv2_D'),
        'W_fc1': tf.Variable(tf.truncated_normal([7*7*64, num_classes], stddev=0.1), name='W_fc1_D'),
    }

biases_D = {
    'b_conv1': tf.Variable(tf.constant(0.1, shape=[32]), name='b_conv1_D'),
    'b_conv2': tf.Variable(tf.constant(0.1, shape=[64]), name='b_conv2_D'),
    'b_fc1': tf.Variable(tf.constant(0.1, shape=[num_classes]), name='b_fc1_D'),             
}

# filter   conv2d conv2d_transpose  
weights_G = {            
    'W_conv1': tf.Variable(tf.truncated_normal([5, 5, 32, 64], stddev=0.1), name='W_conv1_G'), 
    'W_conv2': tf.Variable(tf.truncated_normal([5, 5, 1, 32], stddev=0.1), name='W_conv2_G'),            
#     'W_conv2': tf.Variable(tf.truncated_normal([5, 5, 32, 1], stddev=0.1), name='W_conv2_G'),                        
}

biases_G = {
#     'b_fc1': tf.Variable(tf.constant(0.1, shape=[64]), name='b_fc1_G'),
    'b_conv1': tf.Variable(tf.constant(0.1, shape=[32]), name='b_conv1_G'),
    'b_conv2': tf.Variable(tf.constant(0.1, shape=[1]), name='b_conv2_G')                         
}
```

```python
def discriminator(x_input, num_classes, weights_D, biases_D):
   
        
    # model
    with tf.name_scope('input_reshape_D'):
        x_input #= tf.placeholder(tf.float32, [None, self.input_size])
        x_image = tf.reshape(x_input, [-1,28,28,1], name="x_image_reshape_D") # (Batch_size, width, height, Dim of Channel) .

    with tf.name_scope("Conv_Layer1_D"):
        conv1 = tf.nn.conv2d(x_image, weights_D['W_conv1'], strides=[1, 2, 2, 1], padding='SAME') 
        h_conv1 = lrelu(conv1 + biases_D['b_conv1'])

    with tf.name_scope("Conv_Layer2_D"):
        conv2 = tf.nn.conv2d(h_conv1, weights_D['W_conv2'], strides=[1, 2, 2, 1], padding='SAME')
        h_conv2 = lrelu(conv2 + biases_D['b_conv2'])

    with tf.name_scope("FC_Layer_D"):
        h_conv2_flat = tf.reshape(h_conv2, [-1, 7*7*64], name="h_pool2_flat")
        outputs = tf.tanh(tf.matmul(h_conv2_flat, weights_D['W_fc1']) + biases_D['b_fc1'], name="h_fc1_D")
#         pre_activate_fc = tf.nn.conv2d(conv2, weights_D['W_fc1'], strides=[1, 1, 1, 1], padding='SAME') + biases_D['b_fc1'] #softmax                         
#         outputs = tf.tanh(pre_activate_fc)
        
    return outputs

        
```

```python
def generator(x_input, weights_G, biases_G):
    # output shape ...
    # model
    with tf.name_scope('input_reshape_G'):            
        x_input# = tf.placeholder(tf.float32, [None, self.input_size])
        x_noise = tf.reshape(x_input, [-1,7,7,64], name="x_noise_reshape_G") # (Batch_size, width, height, Dim of Channel) .

    with tf.name_scope("Conv_Layer1_G"):
        fc = tf.nn.conv2d_transpose(x_noise, weights_G['W_conv1'], output_shape=[batch_size, 14, 14, 32], strides=[1, 2, 2, 1], padding='SAME')  
        h_fc1 = lrelu(fc + biases_G['b_conv1'])

    with tf.name_scope("Conv_Layer2_G"):
        conv1 = tf.nn.conv2d_transpose(h_fc1, weights_G['W_conv2'], output_shape=[batch_size, 28, 28, 1],strides=[1, 2, 2, 1], padding='SAME') 
        h_conv1 = lrelu(conv1 + biases_G['b_conv2'])

#     with tf.name_scope("Conv_Layer2_G"):
#         conv2 = tf.nn.conv2d_transpose(h_conv1, weights_G['W_conv2'], strides=[1, 2, 2, 1], padding='SAME')
#         h_conv2 = tf.tanh(conv2 + biases_G['b_conv2']) # Sigmoid  
    
    fake_images = h_conv1 #h_conv2
    
    return fake_images
                        
```

```python


sess = tf.Session()

x_inputs = tf.placeholder(tf.float32, [None, input_size], name='real_images')
z_inputs = tf.placeholder(tf.float32, [None, z_size], name='latent_z_vector')

# Train Discriminator 
# index 0  True, 1  False  .
outputs_D = discriminator(x_inputs, num_classes, weights_D, biases_D)
real_labels = tf.one_hot(indices = tf.zeros(batch_size, dtype=tf.int32), depth = num_classes)
print(outputs_D)
print(real_labels)
real_loss = tf.reduce_sum(tf.pow((outputs_D - real_labels), 2.0))


fake_images = generator(z_inputs, weights_G, biases_G)
outputs_G = discriminator(fake_images, num_classes, weights_D, biases_D)
fake_labels = tf.one_hot(indices = tf.ones(batch_size, dtype=tf.int32), depth = num_classes)
fake_loss = tf.reduce_sum(tf.pow((outputs_G - fake_labels), 2.0))
d_loss = real_loss + fake_loss
train_op_D = tf.train.AdamOptimizer(learning_rate).minimize(d_loss)

# Train Generator 
fake_images = generator(z_inputs, weights_G, biases_G)
outputs_G = discriminator(fake_images, num_classes, weights_D, biases_D)
g_loss = tf.reduce_sum(tf.pow((outputs_G - real_labels), 2.0))
train_op_G = tf.train.AdamOptimizer(learning_rate).minimize(g_loss)

init = tf.global_variables_initializer()
sess.run(init)
```

```python
# ToDo: Training .. def   Dynamic Graph 
correct_prediction = tf.equal(tf.argmax(outputs_D,1), tf.argmax(real_labels,1))
accurcay = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
Flag_for_D = 0
accurcay_out = 0
d_loss_out = 0
g_loss_out = 0
for i in range(num_of_iter):
    batch = mnist.train.next_batch(batch_size)
    batch_image_flat = batch[0].reshape(-1, 28*28)
    batch_input_noise = np.random.uniform(-2,2,(batch_size,7*7*64))
    batch_input_noise = batch_input_noise.reshape(-1, 7*7*64)
#     print(np.shape(batch_input_noise))
    
    if(accurcay_out >= 0.99 and d_loss_out < 201 and g_loss_out < 101 and Flag_for_D == 0):
        Flag_for_D = 1
        print("Flag_for_D = 1")
    
    if not(Flag_for_D == 1):
        _, d_loss_out, accurcay_out = sess.run([train_op_D, d_loss, accurcay], 
                                           feed_dict={x_inputs: batch_image_flat,
                                                     z_inputs: batch_input_noise})

    
    _, g_loss_out, fake_images_out = sess.run([train_op_G, g_loss, fake_images], 
                                   feed_dict={x_inputs: batch_image_flat,
                                             z_inputs: batch_input_noise})
    
    
    
    
    
    if i%100 == 0:
        print("step: %d, d_loss: %g, g_loss: %g, accurcay: %g"%(i, d_loss_out, g_loss_out, accurcay_out))
        classes = np.argmax(batch[1], 1)
        print("Original Image (step: %d)"%(i))
        plot_mnist(batch[0],classes, 1)
        print("")
        print("Generated Image (step: %d)"%(i))
        fake_images_out = fake_images_out.reshape(-1, 28, 28)
        plot_mnist(fake_images_out,classes, 1)
#         train_accuracy = accuracy.eval(feed_dict={x:batch[0], y_: batch[1], keep_prob: keep_prob_test})
```
